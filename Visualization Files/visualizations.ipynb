```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

# Sample data (replace this with your actual data)
data = pd.read_csv('businessEntities.csv') 


# Creating a DataFrame from the data
df = pd.DataFrame(data)

# Bar graph for businessType counts
plt.figure(figsize=(8, 6))
business_type_counts = df['businessType'].value_counts()
business_type_counts.plot(kind='bar')
plt.title('Business Type Counts')
plt.xlabel('Business Type')
plt.ylabel('Count')
plt.show()

# Read the shapefile for the entire United States
gdf = gpd.read_file('cb_2018_us_state_5m.shp')

# Replace 'state' with the actual column name in your data that contains state acronyms
state_acronyms = df['state'].value_counts()

# Create a new column in gdf to store state_acronyms counts
gdf['state_counts'] = gdf['STATE_ABBR'].map(state_acronyms)

# Plot the heatmap
plt.figure(figsize=(10, 8))
gdf.plot(column='state_counts', cmap='Blues', linewidth=0.8, edgecolor='0.8', legend=True)
plt.title('Heatmap of States based on State Acronyms')
plt.show()
```
